```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

**Load the data**

```python
file_path = '/content/all_data.csv'
data = pd.read_csv(file_path)
```

**Print summery of data**

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

**Unique Countries in the dataset**

```python
# Insights
print("\nUnique Countries in the Dataset:")
print(data['Country'].unique())
```

```python
print("\nUnique Years in the Dataset:")
print(data['Year'].unique())
```

```python
# Has life expectancy increased over time in the six nations?
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Year', y='Life expectancy at birth (years)', hue='Country', marker='o')
plt.title('Life Expectancy Over Time in Six Nations')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```

```python
# Has GDP increased over time in the six nations?
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x='Year', y='GDP', hue='Country', marker='o')
plt.title('GDP Over Time in Six Nations')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```

```python
# correlation between GDP and life expectancy
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='GDP', y='Life expectancy at birth (years)', hue='Country', alpha=0.7)
plt.title('Correlation Between GDP and Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```

```python
# average life expectancy
average_life_expectancy = data.groupby('Country')['Life expectancy at birth (years)'].mean()
print("\nAverage Life Expectancy by Country:")
print(average_life_expectancy)
```

**Data Visualization Findings**

1. **Life Expectancy Trends**:
   - All six nations exhibit an increase in life expectancy over the years (2000–2015).
   - Developed countries like Germany and the USA maintain high life expectancies, consistently above 75 years.
   - Developing nations like Zimbabwe show substantial improvements, rising from around 45 years in 2000 to over 60 years in 2015.

2. **GDP Trends**:
   - The United States leads in GDP, starting at $10 trillion in 2000 and reaching nearly $17 trillion by 2015.
   - China's rapid GDP growth reflects its economic transformation into a global powerhouse, with GDP surging dramatically during the time frame.
   - Zimbabwe has the lowest GDP with minimal growth, yet its life expectancy shows significant improvement.

3. **Correlation Between GDP and Life Expectancy**:
   - The scatterplot clearly shows a **positive correlation** between GDP and life expectancy.
   - Countries with higher GDP, such as the USA and Germany, consistently exhibit higher life expectancies.
   - Developing countries like Zimbabwe, while having a low GDP, show a significant increase in life expectancy, indicating improvements in healthcare and social infrastructure.
   - Chile demonstrates an outlier trend where a moderate GDP supports a relatively high life expectancy, suggesting the influence of factors beyond GDP, such as healthcare accessibility and social policies.

4. **Insights from the Scatterplot**:
   - **Linear Correlation**: Developed countries like the USA and Germany follow a trend where GDP increases align with higher life expectancy.
   - **Cluster Analysis**:
     - High-GDP countries form a cluster with life expectancy ranging from 75 to 80 years.
     - Low-GDP nations like Zimbabwe form a separate cluster with life expectancy improving from 45 to 60 years.
   - **Exceptions**:
     - Mexico and Chile maintain high life expectancies despite their GDP being significantly lower than that of the USA or Germany, indicating effective healthcare and social strategies.
     - The USA, despite having the highest GDP, does not surpass other developed nations in life expectancy, reflecting disparities in health outcomes.

5. **Key Takeaways**:
   - While GDP is a strong indicator of life expectancy, it is not the sole determinant. Social policies, healthcare quality, education, and equity play pivotal roles in shaping life expectancy.
   - The scatterplot emphasizes the importance of addressing disparities within high-GDP countries like the USA and leveraging effective policies in developing nations.

This analysis, supported by the scatterplot visualization, underscores the multifaceted relationship between economic growth and public health outcomes.
